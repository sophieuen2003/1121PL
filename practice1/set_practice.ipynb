```python
# 引入 pandas 套件
import pandas as pd
```

```python
# 設定第一個 CSV 檔案(Spotify Million Song Dataset)的路徑
csv_file_path_spmi = 'spotify_millsongdata.csv'

# 使用 pandas 的 read_csv 函式讀取 CSV 檔案並將其轉換為 DataFrame 並使用 utf-8 的編碼方式來處理這個檔案
df_spmi = pd.read_csv(csv_file_path_spmi, encoding = "utf-8")


# 印出 CSV 檔裡面的所有資料，來檢視和原始 CSV 檔有無出入
df_spmi
```

```python
# 設定第二個 CSV 檔案(spotify top chart songs 2022)的路徑
csv_file_path_sp22 = 'spotify_top_charts_22.csv'


# 使用 pandas 的 read_csv 函式讀取 CSV 檔案並將其轉換為 DataFrame 並使用 utf-8 的編碼方式來處理這個檔案
df_sp22 = pd.read_csv(csv_file_path_sp22, encoding = "utf-8")


# 印出 CSV 檔裡面的所有資料，來檢視和原始 CSV 檔有無出入
df_sp22
```

```python
# 從 DataFrame 中取出歌曲名稱欄位
music_names_spmi = df_spmi['song']
music_names_sp22 = df_sp22['track_name']


# 將歌曲名稱轉換為 set 的資料型態
music_names_spmi_set = set(music_names_spmi)
music_names_sp22_set = set(music_names_sp22)


# 印出同時出現在 Spotify 破百萬歌曲和2022年排行榜的歌曲
print(music_names_spmi_set & music_names_sp22_set)
```

```python
# 印出出現在 Spotify 破百萬歌曲但沒出現在2022年排行榜的歌曲
print(music_names_spmi_set - music_names_sp22_set)
```

```python
# 印出出現在 Spotify 2022年排行榜但沒破百萬的歌曲
print(music_names_sp22_set - music_names_spmi_set)
```
