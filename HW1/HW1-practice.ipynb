```python
import pandas as pd
data = pd.read_csv('adult.data.csv', header=None)
data.head()
```

```python
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num',
              'marital_status', 'occupation', 'relationship', 'race', 'sex',
              'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
```

```python
data.head()
```

```python
#哪些國家有更高的薪水？哪些國家薪水比較低？

# 計算不同國家中高收入和低收入的人數
income_by_job = data.groupby(['native_country', 'income'])['sex'].count().unstack()
income_by_job
```

```python
#受教育時間是否影響收入？各國的受教育時間平均
hours_per_week_by_job = data.groupby(['native_country', 'income'])['education_num'].agg(['max', 'min', 'mean', 'std', 'median'])

hours_per_week_by_job
```

```python
income_age = data[['income', 'education_num']]

import plotly.express as px

fig = px.violin(income_age, y="education_num", x="income", color="income", 
                box=True, points="all",
          hover_data=income_age.columns)
fig.show()
```

```python
#哪個性別有更高的薪水？

# 計算不同國家中高收入和低收入的人數
income_by_job = data.groupby(['sex', 'income'])['sex'].count().unstack()
income_by_job
```

```python
#哪個性別的工作時長較長？

# 計算不同國家中高收入和低收入的人數
income_by_job = data.groupby(['sex', 'hours_per_week'])['sex'].count().unstack()
income_by_job
```
