想解決的問題：
1. 臺灣加權指數和時間的關係
2. 美國道瓊工業指數和時間的關係
3. 臺灣加權指數和美國道瓊工業指數的關係
4. 查看各指數的折線圖

```python
# 引入套件
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 設定 CSV 檔案(stock_index)的路徑
csv_file_path_spin = 'stock_index.csv'

# 使用 pandas 的 read_csv 函式讀取 CSV 檔案並將其轉換為 DataFrame 並使用 utf-8 的編碼方式來處理這個檔案
df_stin = pd.read_csv(csv_file_path_spin, encoding = 'utf-8')

df_stin
```

```python
# 儲存臺灣加權指數的數據
TAIEX = df_stin['臺灣加權指數']

TAIEX
```

```python
# 設定圖表標題(臺灣加權指數和時間的關係)

# 設定X軸及Y軸
X = df_stin['日期']
Y = df_stin['臺灣加權指數']

# 設定圖表外觀
fig, ax = plt.subplots(figsize=(12, 6))
x_ticks = np.arange(0, len(X), 100000)
ax.tick_params(axis='x', labelsize=8 ,rotation=90)


ax.plot(X, Y)
fig.show()
```

```python
# 儲存美國道瓊工業指數的數據
DJI = df_stin['美國道瓊工業指數']

DJI
```

```python
# 設定圖表標題(美國道瓊工業指數和時間的關係)

# 設定X軸及Y軸
X = df_stin['日期']
Y = df_stin['美國道瓊工業指數']

# 設定圖表外觀
fig, ax = plt.subplots(figsize=(12, 6))
x_ticks = np.arange(0, len(X), 100000)
ax.tick_params(axis='x', labelsize=8 ,rotation=90)


ax.plot(X, Y)
fig.show()
```

```python
# 圖表設定(臺灣加權指數和美國道瓊工業指數的關係)

X=df_stin['臺灣加權指數']
Y=df_stin['美國道瓊工業指數']

fig, ax = plt.subplots(figsize=(12, 6))

ax.scatter(X, Y)
fig.show()
```
